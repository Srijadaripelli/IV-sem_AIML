# Visualizing Data-Multivariate Plots in Python Machine Learning

```python
#Correlation Matrix Plot
# feature to feature comparison with combinations
#close to 0 independent, +1 strong positively correlated, -1 negatively correlated
fig=plt.figure()
ax=fig.add_subplot(111)
correlations=df.corr()
print(correlations)
cax=ax.matshow(correlations,vmin=-1,vmax=1)
fig.colorbar(cax)

ticks=np.arange(0,13,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
plt.show()
```

# Rescaling Data

```python
# create an object to scale and then transform
#import numpy as np
dfarray = df.values
print(dfarray.shape, type(dfarray))

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-1,1))
rescaledX=scaler.fit_transform(dfarray)
pdscaled = pd.DataFrame(rescaledX)
print(pdscaled.shape)
print(pdscaled.head(10))
pdscaled.describe()
```

# Standardizing Data

```python
# for every value x-mean/deviation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(dfarray)
stdrescaledX=scaler.transform(dfarray)
pdstdscaled = pd.DataFrame(stdrescaledX)
print(pdstdscaled.shape)
print(pdstdscaled.head(10))
pdstdscaled.describe()
```

```python
pdstdscaled.hist()
plt.show()
```

# Normalizing Data

```python
from sklearn.preprocessing import Normalizer
scaler=Normalizer().fit(dfarray)
normalizedX=scaler.transform(dfarray)

pdnormalscaled = pd.DataFrame(normalizedX)
print(pdnormalscaled.shape)
print(pdnormalscaled.head(10))
pdnormalscaled.describe()
```

```python
npvalues= pdnormalscaled[111:112].values
npvalues.sum()
#npvalues
```

# Binarizing Data

```python
from sklearn.preprocessing import Binarizer
binarizer=Binarizer(threshold=0.0).fit(dfarray)
binaryX=binarizer.transform(dfarray)

pdbinarized = pd.DataFrame(binaryX)
print(pdbinarized.shape)
print(pdbinarized.tail(10))
```
